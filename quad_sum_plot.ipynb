```python
import pandas as pd
df = pd.read_csv("quadratic-sums-result.csv")
df
```

```python
import matplotlib.pyplot as plt

ax = df.plot(
    x='l-sample',
    y=['multiply-shift-ms', 'multiply-mod-prime-ms'],
    marker='o'
)
ax.set_xlabel('l-sample')
ax.set_ylabel('Time (ms)')
ax.set_title('multiply-shift vs multiply-mod-prime')
ax.grid(True)
```

```python
percent_diff = (df['multiply-mod-prime-ms'] - df['multiply-shift-ms']) / df['multiply-shift-ms'] * 100

fig, ax2 = plt.subplots()
ax2.plot(df['l-sample'], percent_diff, marker='o')
ax2.axhline(0, color='gray', linewidth=0.8)
ax2.set_xlabel('l-sample')
ax2.set_ylabel('Percent difference (%)')
ax2.set_title('Percent-wise difference between multiply-shift-ms and multiply-mod-prime-ms')
ax2.grid(True)
```
