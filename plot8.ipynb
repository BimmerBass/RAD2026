```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("results_opg8.csv")
```

```python
# Filter first 100 rows and plot
first_100 = df.head(100)

fig, ax = plt.subplots(figsize=(10, 5))


ax.plot(first_100['sorted_value'], label='sorted_value', marker='o', linestyle='-',)
ax.axhline(y=exact, color='r', linestyle='--', label='exact')
ax.set_xlabel('Row Index')
ax.set_ylabel('Value')
ax.set_title('Sorted f2 Estimates by Index, with bitwidth 8')
ax.legend()
plt.show()
```

```python
segment = df.head(100)

l = segment['counter_bitwidth'].iloc[0]
num_buckets = 1 << l
exact_f2_value = segment['exact'].iloc[0]
mse_value = segment['mse'].mean()

exp_variance = 2 * exact_f2_value**2 / num_buckets

print(f"Expected variance of the estimator: {exp_variance:.6g}")
print(f"Empirical variance of the estimator: {mse_value:.6g}")
print(f"Error (|empirical variance - expected variance|): {abs(mse_value - exp_variance):.6g} ({abs(mse_value - exp_variance) / exp_variance * 100:.6f}%)")
```
